```python
train[['release_month', 'release_day', 'release_year']] = train['release_date'].str.split('/', expand=True).replace(np.nan, -1).astype(int)
train['release_month'].describe()
```

```python
train['release_day'].describe()
```

```python
train['release_year'].describe()
```

```python
train.loc[ (train['release_year'] <= 19) & (train['release_year'] >= 0), 'release_year'] += 2000
train.loc[ (train['release_year']  > 19) & (train['release_year'] < 100), 'release_year'] += 1900
train['release_year'].head(20)
```

```python
plt.figure(figsize=(20, 10))
sns.countplot(x=train['release_year'].sort_values())
plt.title("Movie releases on each year",fontsize=20)
loc, labels = plt.xticks()
plt.xticks(fontsize=14,rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.countplot(x=train['release_month'].sort_values())
plt.title('Movie releases on each month',fontsize=20)
loc, labels = plt.xticks()
loc, labels = loc, ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
plt.xticks(loc, labels,fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.countplot(x=train['release_day'].sort_values())
plt.title('Movie releases on each day',fontsize=20)
plt.show()
```

```python
release_date = pd.to_datetime(train['release_date'])
train['release_day_week'] = release_date.dt.dayofweek
train['release_quarter'] = release_date.dt.quarter
```

```python
plt.figure(figsize=(20, 10))
sns.countplot(x=train['release_day_week'].sort_values())
plt.title('Movie releases on each day of week', fontsize=20)
loc, labels = plt.xticks()
loc, labels = loc, ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
plt.xticks(loc, labels,fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.countplot(x=train['release_quarter'].sort_values())
plt.title('Movie releases on each quarter', fontsize=20)
plt.show()
```
