```python
from catboost import CatBoostRegressor
```

```python
x = train.drop('revenue', axis=1)
y = train['revenue']
x.drop('genres', axis=1, inplace=True)
```

```python
x['original_language'] = x['original_language'].astype(str)
x['status'] = x['status'].astype(str)
x['release_date'] = x['release_date'].astype(str)
cat_features = ['original_language', 'status', 'has_homepage', 'has_tagline', 'is_original_english', 'release_date']
```

```python
x.drop('log_revenue', axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 52)
```

```python
model = CatBoostRegressor(verbose=0, depth=6, learning_rate=0.1)
model.fit(x_train, y_train, cat_features=cat_features)
```

```python
x
```

```python
from sklearn.metrics import mean_squared_log_error
y_pred =  model.predict(x_test)
y_pred = np.clip(y_pred, 0, None)
rmsle = np.sqrt(mean_squared_log_error(y_test, y_pred))
print(f'Root Mean Squared Logarithmic Error (RMSLE): {rmsle}')
```
